```python
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
#FUNCTIONAL METHOD
plt.plot(x,y)
plt.xlabel('X -->')
plt.ylabel('y -->')
plt.title('Tit le lo')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'r-')
```

```python
plt.subplot(1,2,2)
plt.plot(y,x,'g-')
```

```python
#OBJECT ORIENTED METHOD
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('x--')
axes.set_ylabel('y--')
axes.set_title('Title is random')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,0.8,0.8])
axes2 = fig.add_axes([0.1,0.4,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')
axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

```python
fig,axes = plt.subplots(nrows = 1, ncols= 2)
plt.tight_layout()
#axes.plot(x,y)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows = 1, ncols= 2)
plt.tight_layout()
#axes.plot(x,y)

axes[0].plot(x,y)
axes[1].plot(y,x)
```

FIGURE SIZE AND DPI

```python
fig = plt.figure(figsize=(6,2), dpi = 400)

ax = fig.add_axes([0,0,1,1])

ax.set_xlim([0, max(x)])
ax.set_ylim([0, max(y)])

ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows = 2, ncols = 1, figsize = (3,4))
plt.tight_layout()

axes[0].plot(x,y)
axes[1].plot(y,x)
```
